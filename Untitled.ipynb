```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_cf = pd.read_csv('../Homosapiens_GRCh38_87_gene_length.tab', sep=' ', index_col=0)
df_sarb = pd.read_csv('../sarba_genelength.csv', sep='\t', index_col=1)
df_sarb.drop('Gene', axis=1)
```

```python
df_cf
```

```python
df_comb = pd.DataFrame()
df_comb['Length_old'] = df_sarb['Length']
df_comb['Length_new'] = df_cf['Length']
df_comb
```

```python
len(df_cf)
```

```python
len(df_sarb)
```

```python
from scipy import stats
def r2(x, y):
    return stats.pearsonr(x, y)[0] ** 2
sns.jointplot(df_comb['Length_new'], df_comb['Length_old'], kind="reg", stat_func=r2)
```

```python
#fig = plt.figure()
sns.lmplot('Length_new', 'Length_old', data=df_comb)
#fig.legend()
```
